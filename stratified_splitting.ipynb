#### split the imbalance dataset into subsets which have proportional representation

```python
from sklearn.datasets import make_multilabel_classification
from skmultilearn.model_selection.iterative_stratification import IterativeStratification
import numpy as np
import pandas as pd


# Assuming you have your features X and labels y
# Generate a random multilabel classification dataset
X, y = make_multilabel_classification(n_samples=1000, n_features=20, n_classes=5, random_state=42)

# Specify the number of splits (folds) you want for cross-validation
n_splits = 5

classes = ['first','second','third','forth','fifth']

# Create an instance of IterativeStratification with the desired parameters
iterative_stratifier = IterativeStratification(n_splits=n_splits, order=1)

# Use the iterative_stratifier to generate indices for cross-validation
for train_index, test_index in iterative_stratifier.split(X, y):
    # Your training and testing data
    X_train, X_test = X[train_index], X[test_index]
    print(X_train.shape, X_test.shape)
    
    y_train, y_test = y[train_index], y[test_index]
    print(pd.Series(np.stack(y_train, axis = 0).sum(axis=0), index=classes))
    print(pd.Series(np.stack(y_test, axis = 0).sum(axis=0), index=classes))
    print(y_train.shape, y_test.shape)
    print("------------------------------------------")
    # Now you can train and evaluate your model using the current split
    # ...
```
